```python
import pandas as pd
import time
from openai import OpenAI
import os
```

```python
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
)

api_model_name = "openai/gpt-4o-mini"
```

```python
def translate_arabic_to_english(text):

    text = str(text)

    response = client.chat.completions.create(
        model=api_model_name,
        messages=[
            {
                "role": "system",
                "content": "You are a professional Arabic to English translator. Translate the Arabic text into English only. Do not add explanation."
            },
            {
                "role": "user",
                "content": "Translate this Arabic text into English:\n" + text
            }
        ],
        max_tokens=500,
        temperature=0.2
    )

    translated_text = response.choices[0].message.content

    return translated_text.strip()
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df['question_text'] = (
    df['QuestionText']
    .astype(str)
    .str.replace("[", "", regex=False)
    .str.replace("]", "", regex=False)
    .str.replace("'", "", regex=False)
    .str.replace(",", " ", regex=False)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

df['answer_text'] = (
    df['Answer']
    .astype(str)
    .str.replace("[", "", regex=False)
    .str.replace("]", "", regex=False)
    .str.replace("'", "", regex=False)
    .str.replace(",", " ", regex=False)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

df[['question_text', 'answer_text']].head()
```

```python
for i in range(5):

    sample_question = df['question_text'].iloc[i]
    sample_answer = df['answer_text'].iloc[i]

    print("\n==============================")
    print("Example:", i + 1)

    print("\nArabic Question:")
    print(sample_question)

    print("\nEnglish Question:")
    print(translate_arabic_to_english(sample_question))

    print("\nArabic Answer:")
    print(sample_answer)

    print("\nEnglish Answer:")
    print(translate_arabic_to_english(sample_answer))
```
